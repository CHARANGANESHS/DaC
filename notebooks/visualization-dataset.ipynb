```python
from medmnist import OrganSMNIST, OrganAMNIST, OrganCMNIST
import matplotlib.pyplot as plt
import numpy as np
```

```python
def visualize_same_label(dataset, label, title, num_samples=5):
    filtered_images = [dataset[i][0] for i in range(len(dataset)) if dataset[i][1] == label]
    fig, axes = plt.subplots(1, min(num_samples, len(filtered_images)), figsize=(15, 5))
    fig.suptitle(title, fontsize=16)
    for i in range(min(num_samples, len(filtered_images))):
        image = filtered_images[i]
        if not isinstance(image, np.ndarray):
            image = np.array(image)
        axes[i].imshow(image.squeeze(), cmap='gray')
        axes[i].axis('off')
        axes[i].set_title(f"Label: {label}")
    plt.show()

amnist = OrganAMNIST(split='train', download=True)
cmnist = OrganCMNIST(split='train', download=True)

visualize_same_label(amnist, label=3, title="OrganAMNIST - Label 3 (Heart)")
visualize_same_label(cmnist, label=3, title="OrganCMNIST - Label 3 (Heart)")
```
